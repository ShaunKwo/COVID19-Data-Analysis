Extracting Singapore data from COVID-19 Cases and deaths - WHO.csv

```python
import pandas as pd

covid_death_cases=pd.read_csv('./PDAS raw data/COVID-19 Cases and deaths - WHO.csv')
print(covid_death_cases)
```

```python
# Filter rows where the 'Country' column is 'Singapore'
singapore_covid_death_cases = covid_death_cases[covid_death_cases['Country'] == 'Singapore']
missing=singapore_covid_death_cases.isnull().sum()
print(missing)

# Display the resulting dataframe
print(singapore_covid_death_cases)


# imputing by mode
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent') # (mode)

# Fit and transform the DataFrame to impute missing values
singapore_covid_death_cases_imputed = pd.DataFrame(imputer.fit_transform(singapore_covid_death_cases), columns=singapore_covid_death_cases.columns)

# Print the imputed DataFrame
print("Original DataFrame:")
print(singapore_covid_death_cases)
print("\nImputed DataFrame:")
print(singapore_covid_death_cases_imputed)

# Save the data to a new Excel file
singapore_covid_death_cases_imputed.to_excel('./extracted_sg_data/singapore_covid_death_cases.xlsx', engine='openpyxl')
```

Extracting Singapore data from real-gdp-growth.csv

```python
real_gdp_growth=pd.read_csv('./PDAS raw data/real-gdp-growth.csv')
missing=real_gdp_growth.isnull().sum()
print(missing)
print(real_gdp_growth)

# imputing by mode
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent') # (mode)

# Fit and transform the DataFrame to impute missing values
real_gdp_growth_imputed = pd.DataFrame(imputer.fit_transform(real_gdp_growth), columns=real_gdp_growth.columns)

# Print the imputed DataFrame
print(real_gdp_growth_imputed.isnull().sum())
print("\nImputed DataFrame:")
print(real_gdp_growth_imputed)
```

```python
# Filter rows where the 'Country' column is 'Singapore'
singapore_real_gdp_growth = real_gdp_growth[real_gdp_growth['Entity'] == 'Singapore']

# Display the resulting dataframe
print(singapore_real_gdp_growth)

# Save the data to a new Excel file
singapore_real_gdp_growth.to_excel('./extracted_sg_data/real_gdp_growth.xlsx', engine='openpyxl')
```
